```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
df.columns
```

```python
# List of categorical variables
categorical_vars = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked']

# Create bar charts for each categorical variable
for var in categorical_vars:
    plt.figure(figsize=(8, 6))
    df[var].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.show()
```

```python
# List of continuous variables
continuous_vars = ['Age', 'Fare']

# Create histograms for each continuous variable
for var in continuous_vars:
    plt.figure(figsize=(8, 6))
    df[var].dropna().plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.show()
```
