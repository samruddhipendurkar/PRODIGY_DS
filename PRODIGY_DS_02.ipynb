```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from a CSV file
df = pd.read_csv('Titanic-Dataset.csv')

# Display the first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Example: Fill missing values or drop rows/columns with missing values
# For simplicity, let's drop rows with missing values
df_cleaned = df.dropna()

# Alternatively, you can fill missing values with a specific value or strategy
# df_cleaned = df.fillna(method='ffill')  # Forward fill
```

```python
# Check for duplicate rows
print(df_cleaned.duplicated().sum())

# Remove duplicate rows
df_cleaned = df_cleaned.drop_duplicates()
```

```python
# Summary statistics
print(df_cleaned.describe(include='all'))
```

```python
# Distribution of 'Survived'
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df_cleaned, palette='Set2')
plt.title('Distribution of Survived Variable')
plt.show()
```

```python
# Example: Survived vs. Pclass (Passenger Class)
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df_cleaned, palette='Set2')
plt.title('Survived vs. Pclass')
plt.show()
```

```python
# Example: Survived vs. Sex
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=df_cleaned, palette='Set2')
plt.title('Survived vs. Sex')
plt.show()
```

```python
# Correlation matrix to identify relationships between numerical variables
# Select only numerical columns for correlation calculation
numerical_columns = df_cleaned.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
