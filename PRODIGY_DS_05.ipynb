```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('Traffic_Collisions_Toronto_data.csv')
df.head()
```

```python
# Drop rows with missing values in critical columns
df.dropna(subset=['OccurrenceDate', 'Hour', 'Longitude', 'Latitude'], inplace=True)

# Convert 'OccurrenceDate' to datetime
df['OccurrenceDate'] = pd.to_datetime(df['OccurrenceDate'])

# Extract day of week and hour from 'OccurrenceDate'
df['DayOfWeek'] = df['OccurrenceDate'].dt.day_name()
df['Hour'] = df['Hour'].astype(int)

# Display first few rows to verify changes
print(df.head())
```

```python
# Plot distribution of injury collisions
plt.figure(figsize=(10, 6))
sns.countplot(x='Injury_Collisions', data=df, palette='Set2')
plt.title('Distribution of Injury Collisions')
plt.xlabel('Injury Collisions')
plt.ylabel('Count')
plt.show()
```

```python
# Plot distribution of FTR collisions
plt.figure(figsize=(10, 6))
sns.countplot(x='FTR_Collisions', data=df, palette='Set2')
plt.title('Distribution of FTR Collisions')
plt.xlabel('FTR Collisions')
plt.ylabel('Count')
plt.show()
```

```python
# Plot distribution of PD collisions
plt.figure(figsize=(10, 6))
sns.countplot(x='PD_Collisions', data=df, palette='Set2')
plt.title('Distribution of Property Damage (PD) Collisions')
plt.xlabel('PD Collisions')
plt.ylabel('Count')
plt.show()
```

```python
# Accidents by hour of the day
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=df, palette='Set2')
plt.title('Distribution of Accidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.show()
```
